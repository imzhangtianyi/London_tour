```python
from bs4 import BeautifulSoup
import requests
import time
import psycopg2 as pg2
```

```python
page_response = requests.get('https://www.tripadvisor.com/Attraction_Review-g186338-d7224941-Reviews-The_Queen_s_Walk-London_England.html',
                                 timeout=5)
page_content = BeautifulSoup(page_response.content, "html.parser")
```

```python
soup = page_content
try:
    attraction = soup.find('h1', {'id':'HEADING'}).getText()
except:
    attraction = 'missing'

try:
    nreview = soup.find('a', {'href':'#REVIEWS'}).getText().split()[0]
except:
    nreview = 'missing'
try:
    address = soup.find('div', {'class':'detail_section address'}).getText()
except:
    address = 'missing'

try:
    ranking = soup.find('b', {'class':''}).getText()
except:
    ranking = 'missing'

try:
    cate = soup.find('div', {'class':'detail'}).getText()
except:
    cate = 'missing'

try:
    ph = soup.find('div', {'class':'detail_section phone'}).getText()
except:
    ph = 'missing'

try:
    hrs = soup.find('div', {'class':'detail_section hours'}).getText()
except:
    hrs = 'missing'
```
