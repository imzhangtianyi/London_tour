```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
n = range(0,301,30)
u = map(lambda x: 'https://www.tripadvisor.com/Attractions-g186338-Activities-oa' + str(x) + '-London_England.html#FILTERED_LIST', n)
```

```python
i=1
print '}age {}'.format(i)
```

```python
%%time
bl = []
i = 0
for url in u:
    page_response = requests.get(url, timeout=5)
    time.sleep(1)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    b = page_content.find_all('div', {'class':'listing_title'})
    bl.append(b)
    i=i+1
    print 'page {}'.format(i)
```

```python
len(bl)
```

```python
att = map(lambda x: x.find('a').get('href'), bl)
u_att = map(lambda x: 'https://www.tripadvisor.com' + str(x), att)
```
