```python
import pandas as pd
d=pd.read_csv('london_d0.csv',sep='\t')
```

```python
ll = pd.read_csv('latlon.csv',sep='\t')
```

```python
a = d.iloc[:,1:3]
```

```python
def getnum(s):
    return filter(lambda x: x.isdigit(),s)
df = a.set_index('attraction')
df['ranking'] = df.ranking.apply(getnum)
```

```python
df = df.join(ll.iloc[:,1:].set_index('attraction'))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df.latitude,df.longitude,c=df.ranking,marker = '.',cmap='nipy_spectral')
plt.xlim([51.2,51.8])
plt.ylim([-0.7,0.3])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
m = Basemap(llcrnrlon=40,llcrnrlat=-10,urcrnrlon=10.,urcrnrlat=60.,
             resolution='l', projection='tmerc', lat_0 = 49., lon_0 = -2.)

m.drawcountries()
m.drawcoastlines()
m.fillcontinents(color='#C0C0C0')
plt.show()
```
