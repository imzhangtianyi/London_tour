```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
d=pd.read_pickle('destin_1')
ll = pd.read_csv('latlon.csv',sep='\t')
ll.set_index('attraction',inplace=True)
def getnum(s):
    return filter(lambda x: x.isdigit(),s)
```

```python
a = d.iloc[:,1:3]
df = a.join(ll.iloc[:,1:])
```

```python
df.head()
```

```python
rn = df.loc[df.review_num != 'missing', ['review_num','latitude','longitude']]
```

```python
rn = df.loc[(df.rating != 'missing') & (df.rating != ' '), ['rating','latitude','longitude']]
```

```python
p = rn.rating.apply(getnum)
```

```python
p.value_counts()
```

```python
plt.figure(figsize=(8,8))
m = Basemap(llcrnrlon=-0.6,llcrnrlat=51.2,urcrnrlon=0.4,urcrnrlat=51.8,
            projection='tmerc',lat_0=50.,lon_0=0,
            resolution ='l',)

m.drawrivers(color='b')
m.readshapefile('C:\Users\Administrator\Documents\Web Scraping\de\London_Ward\London_Ward', 'London_Ward',color='#AEB6BF')
x, y = m(rn.longitude.values, rn.latitude.values)
m.scatter(x, y, c=p.values, marker = '.' ,cmap='winter',vmax=50)
plt.show()
```
